# Pandas (Python)

```python
import pandas as pd

# Define data as a list of dictionaries
data = [
    {'employee': 'A1', 'dates': '2024-01-01', 'status': 'PRESENT'},
    {'employee': 'A1', 'dates': '2024-01-02', 'status': 'PRESENT'},
    {'employee': 'A1', 'dates': '2024-01-03', 'status': 'PRESENT'},
    {'employee': 'A1', 'dates': '2024-01-04', 'status': 'ABSENT'},
    {'employee': 'A1', 'dates': '2024-01-05', 'status': 'PRESENT'},
    {'employee': 'A1', 'dates': '2024-01-06', 'status': 'PRESENT'},
    {'employee': 'A1', 'dates': '2024-01-07', 'status': 'ABSENT'},
    {'employee': 'A1', 'dates': '2024-01-08', 'status': 'ABSENT'},
    {'employee': 'A1', 'dates': '2024-01-09', 'status': 'ABSENT'},
    {'employee': 'A1', 'dates': '2024-01-10', 'status': 'PRESENT'},
    {'employee': 'A2', 'dates': '2024-01-06', 'status': 'PRESENT'},
    {'employee': 'A2', 'dates': '2024-01-07', 'status': 'PRESENT'},
    {'employee': 'A2', 'dates': '2024-01-08', 'status': 'ABSENT'},
    {'employee': 'A2', 'dates': '2024-01-09', 'status': 'PRESENT'},
    {'employee': 'A2', 'dates': '2024-01-10', 'status': 'ABSENT'}
]

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'dates' column to datetime format
df['dates'] = pd.to_datetime(df['dates'] , format="%Y-%m-%d")

# Display DataFrame
print(df)
```

```python
df['group'] = (df['status'] != df['status'].shift()).cumsum()
df
```

```python
g = df.groupby(["employee", "group"]).agg(
        FROM_DATE=('dates', min),
        END_DATE=('dates',max),
        STATUS = ('status', 'first')
    ).reset_index()
g.rename(columns=({"employee":"EMPLOYEE"}),inplace=True)
g[["EMPLOYEE", "FROM_DATE", "END_DATE","STATUS"]]
```

# SQL

```python

create table emp_attendance
(
	employee 	varchar(10),
	dates 		date,
	status 		varchar(20)
);

insert into emp_attendance values('A1', '2024-01-01', 'PRESENT');
insert into emp_attendance values('A1', '2024-01-02', 'PRESENT');
insert into emp_attendance values('A1', '2024-01-03', 'PRESENT');
insert into emp_attendance values('A1', '2024-01-04', 'ABSENT');
insert into emp_attendance values('A1', '2024-01-05', 'PRESENT');
insert into emp_attendance values('A1', '2024-01-06', 'PRESENT');
insert into emp_attendance values('A1', '2024-01-07', 'ABSENT');
insert into emp_attendance values('A1', '2024-01-08', 'ABSENT');
insert into emp_attendance values('A1', '2024-01-09', 'ABSENT');
insert into emp_attendance values('A1', '2024-01-010', 'PRESENT');
insert into emp_attendance values('A2', '2024-01-06', 'PRESENT');
insert into emp_attendance values('A2', '2024-01-07', 'PRESENT');
insert into emp_attendance values('A2', '2024-01-08', 'ABSENT');
insert into emp_attendance values('A2', '2024-01-09', 'PRESENT');
insert into emp_attendance values('A2', '2024-01-010', 'ABSENT');

WITH cte AS (
    SELECT 
        employee, 
        status, 
        dates, 
        ROW_NUMBER() OVER (PARTITION BY employee ORDER BY dates) -
        ROW_NUMBER() OVER (PARTITION BY employee, status ORDER BY dates) AS grp
    FROM 
        emp_attendance
)
SELECT 
    employee,
    MIN(dates) AS FROM_DATE,
    MAX(dates) AS TO_DATE,
    status
FROM 
    cte
GROUP BY 
    employee, status, grp
ORDER BY 
    employee, FROM_DATE;
```
